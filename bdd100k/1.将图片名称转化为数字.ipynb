cn-clip中的微调数据集，图片名称是以数字id表示的：
![cn-clip-imageid](images/cn-clip-image-id.png)

而bdd100k数据集，其图片名称是以字符串表示的:
![bdd100-image-name](images/bdd100-image-name.png)

因此我们需要先将图片名称由原来的字符串转化为数字

另外由于bdd100k的图片和标签的表示关系中，图片是以原始字符串表示的，因此为了后面转换图片id和标签的关系，我们需要在将图片名称转化为对应的数字id时，需要先记录图片名称和数字id之间的关系

## 将图片名称以数字id的形式保存，并记录字符串名称数字id的映射

```python
import os
import json
import shutil


def main(srcDir, dstDir, split):
    imageIdMap = {}
    for dirpath, dirnames, filenames in os.walk(srcDir):
        for i, filepath in enumerate(filenames):
            print("processing: {}, {}".format(i, filepath))

            # 将图片名称以数字的形式保存
            fileName = os.path.join(dirpath, filepath)
            imageId = i + 1
            targetFileName = os.path.join(dstDir, str(imageId) + ".jpg")
            shutil.copy(str(fileName), str(targetFileName))

            # 保存图片字符串名称和数字id名称的映射
            imageName = filepath[:-4]
            imageIdMap[imageName] = imageId

    imageIdMapFileName = "{}_image_id_map.json". format(split)
    with open(imageIdMapFileName, 'w') as f:
        json.dump(imageIdMap, f, indent=4, sort_keys=True)
        f.write('\n')  # 写入换行符，表示新的一行


if __name__ == '__main__':
    srcDir = r'E:\playground\ai\datasets\bdd100k\bdd100k_images\bdd100k\images\100k\train'
    dstDir = r'E:\playground\ai\datasets\bdd100k\bdd100k_images\bdd100k\images\100k\train_id'
    split = "train"
    main(srcDir, dstDir, split)
```
